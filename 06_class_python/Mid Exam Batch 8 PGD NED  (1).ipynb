6-Define Mutable and immutable data types with example.\
**Mutable Data Types:**\
Mutable data types are those that allow modification of their contents.\
This means you can change, add, or remove elements without creating a new object.\
**Examples of Mutable Data Types:**\
Lists: You can change, add, or remove elements

```python
# exmple number one:

my_list = [1, 2, 3]
my_list[0] = 10  # Modifying an element, put '10' in the index of '0'. 
my_list.append(4)  # Adding an element, add 4 in the list
print(my_list)

# example number two:

fruit_list = ["Apple","Banana","Cherries","Mango"]
fruit_list[0] = "Orange"
fruit_list.append("Peach")
print("Total fruit in Basket: ",fruit_list)

# example number three:

data = ["Salman","Kashif","Waquar","Ambreen","Adnan","Noureen"]
data[1] = "Soomal"
data[2] = "Shabana"
data[3] = "Abdul Raheem"
data[4] = "Shazia"
data.append("Abdul Hafiz")
print("All family data:",data)
```

Dictionaries: You can change the values associated with keys.

```python
# Example number one:

my_dict = {'a': 1, 'b': 2}
my_dict['a'] = 10  # Changing a value of first key
my_dict['c'] = 3  # Adding a new key-value pair both together
print(my_dict)

# Example number two:

dict_file = {"Biryani": 250, "Koarma": 450, "Tikke": 850}
dict_file['Biryani'] = 350           # changing a value of first key
dict_file['Cold drink']= 250         # Adding a new key_value pair both together
print("The manue list: ",dict_file)

# Example number three:

dict_one = {"Karachi":'Big City', "Islamabad": 'Capital City',"Lahore":'City of Food'}
dict_two = {"Karachi":'Metropoliton City',"Islamabad":'Administrator City',"Lahore":'Cultural City'}
dict_one.update(dict_two)
print("That is my final dictionary: ",dict_one)  # print is giving updated dictionary
```

Sets: You can add or remove elements

```python
# Example number one:

my_set = {1, 2, 3}
my_set.add(4) 
print("Adding 4",my_set)   # Adding an element
my_set.remove(2)
print("Removing 2",my_set)  # Removing an element
print("Total elements of set",my_set)

# Example number two:

empty_set=set()
print("This is empty set:",empty_set)
empty_set.add(4)          # only one element can be add
print("This is added Set:",empty_set)
empty_set.remove(4)
print("Removed will remove 4 from the set and it is empty set: ",empty_set)
```

**Immutable Data Types**\
Immutable data types are those that cannot be changed after they are created.\
Any modification results in the creation of a new object.\
**Examples of Immutable Data Types:**\
Tuples: Once created, you cannot change their elements.

```python
# Example number one:

my_tuple = (1, 2, 3)
my_tuple[0] = 10  # This will raise a TypeError
print(my_tuple)
```

```python
# Example number two:

my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # This will raise a TypeError
print(my_tuple)
```

Strings: Strings cannot be modified; any change creates a new string.

```python
my_string = "Hello"
# my_string[0] = 'h'  # This will raise a TypeError
print("String: ",my_string)
new_string = my_string.replace("H", "h")  # Creating a new string
print("Only with replace can change the alphabet of string: ",new_string) 
```

Frozensets: Similar to sets, but immutable.

```python
my_frozenset = frozenset([1, 2, 3])
# my_frozenset.add(4)  # This will raise an AttributeError
print("Frozenset will not allow you to add any element in the set: ",my_frozenset)
```

```python
# Tuple of coordinates
coordinates = (12.5, 45.8, 33.1)

# Unpacking the tuple into variables
x, y, z = coordinates

# Print the unpacked values
print("x:", x)
print("y:", y)
print("z:", z)
```
