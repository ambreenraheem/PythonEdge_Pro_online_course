**Example # 01:** First, let's create a dictionary of grocery items and their prices.

```python
grocery_prices = {
    'Apples': 150.0,
    'Milk': 220.0,
    'Bread': 225.0,
    'Eggs': 400.0,
    'Cheese': 350.0,
    'Yogurt': 175.0
}

print("Grocery Prices Dictionary:")
print(grocery_prices)
```

**Convert the `grocery_prices` dictionary to a pandas DataFrame**

```python
import pandas as pd

# Convert the grocery_prices dictionary to a pandas DataFrame
grocery_df = pd.DataFrame(list(grocery_prices.items()), columns=['Item', 'Price'])

print("Grocery Prices DataFrame:")
display(grocery_df)
```

Now, let's write a Python code with an input variable and `if-elif-else` statements to check the price of a grocery item. We'll categorize prices as 'Expensive', 'Moderate', or 'Cheap' based on a few thresholds.

```python
expensive_threshold = 400.0
moderate_threshold = 200.0

# Get grocery item name as input
item_name = input("Enter a grocery item to check its price: ")

# Check if the item exists in the dictionary
if item_name in grocery_prices:
    price = grocery_prices[item_name]
    print(f"\nThe price of {item_name} is: Rs{price:.2f}")

    # Apply if-elif-else statement
    if price >= expensive_threshold:
        print(f"{item_name} is considered expensive.")
    elif price >= moderate_threshold:
        print(f"{item_name} has a moderate price.")
    else:
        print(f"{item_name} is relatively cheap.")
else:
    print(f"\nError: Item '{item_name}' not found in the grocery list.")
```

**Example # 02:** First, let's create a dictionary of students and their marks.

```python
student_marks = {
    'Ali': 85,
    'Saad': 45,
    'Murad Ahmed': 92,
    'Rahim': 60,
    'Ehsan': 38
}

print("Student Marks Dictionary:")
print(student_marks)
```

Now, let's write a Python code with an input variable and `if-elif-else` statements to check a student's performance based on their mark. We'll set a passing mark of 50.

```python
passing_mark = 50

# Get student name as input
student_name = input("Enter student's name to check their mark: ")

# Check if the student exists in the dictionary
if student_name in student_marks:
    mark = student_marks[student_name]
    print(f"\n{student_name}'s mark: {mark}")

    # Apply if-elif-else statement
    if mark >= passing_mark:
        print(f"{student_name} has passed! Congratulations!")
    elif mark >= passing_mark - 10: # Example for a 'needs improvement' category
        print(f"{student_name} is close to passing. They need to improve.")
    else:
        print(f"{student_name} has failed. They need to study harder.")
else:
    print(f"\nError: Student '{student_name}' not found in the dictionary.")
```

Establish the numerical thresholds for each grade category (A+, A, B, C, D) based on common academic standards, which will be used for both individual subject grades and overall grades.

## Calculate and Display Grade for Individual Subject

### Subtask:
Retrieve the mark for the specified student and subject from the dictionary. Use `if-elif-else` statements to determine the corresponding grade (A+, A, B, C, or D) based on the defined thresholds and display the result for the individual subject.


**Reasoning**:
To implement the logic for retrieving the individual subject mark and determining its grade, I will use a code block with conditional statements (if-elif-else) based on the previously defined `student_grades` dictionary, user inputs (`student_name_input`, `subject_name_input`), and grading thresholds.



```python
if student_name_input in student_grades:
    if subject_name_input in student_grades[student_name_input]:
        subject_mark = student_grades[student_name_input][subject_name_input]
        individual_grade = ''

        if subject_mark >= grade_aplus:
            individual_grade = 'A+'
        elif subject_mark >= grade_a:
            individual_grade = 'A'
        elif subject_mark >= grade_b:
            individual_grade = 'B'
        elif subject_mark >= grade_c:
            individual_grade = 'C'
        elif subject_mark >= grade_d:
            individual_grade = 'D'
        else:
            individual_grade = 'F'

        print(f"\n--- Individual Subject Grade ---")
        print(f"Student: {student_name_input}")
        print(f"Subject: {subject_name_input}")
        print(f"Mark: {subject_mark}")
        print(f"Grade: {individual_grade}")
    else:
        print(f"\nError: Subject '{subject_name_input}' not found for {student_name_input}.")
else:
    print(f"\nError: Student '{student_name_input}' not found in the records.")
```

## Calculate and Display Overall Percentage and Grade

### Subtask:

Then, use the previously defined grading thresholds to determine the student's overall grade (A+, A, B, C, D, or F) and display both the overall percentage and overall grade.

**Reasoning**:
To calculate the overall percentage and grade, I need to access the student's marks from the `student_grades` dictionary, sum them up, calculate the total possible marks, and then apply the grading thresholds. A code block is necessary for these calculations and conditional logic.



```python
if student_name_input in student_grades:
    student_subjects = student_grades[student_name_input]
    total_marks_obtained = sum(student_subjects.values())
    num_subjects = len(student_subjects)
    total_possible_marks = num_subjects * 100

    if num_subjects > 0:
        overall_percentage = (total_marks_obtained / total_possible_marks) * 100
        overall_grade = ''

        if overall_percentage >= grade_aplus:
           overall_grade = 'A+'
        elif overall_percentage >= grade_a:
           overall_grade = 'A'
        elif overall_percentage >= grade_b:
           overall_grade = 'B'
        elif overall_percentage >= grade_c:
           overall_grade = 'C'
        elif overall_percentage >= grade_d:
           overall_grade = 'D'
        else:
           overall_grade = 'F'

        print(f"\n--- Overall Performance ---")
        print(f"Student: {student_name_input}")
        print(f"Total Marks Obtained: {total_marks_obtained}")
        print(f"Overall Percentage: {overall_percentage:.2f}%")
        print(f"Overall Grade: {overall_grade}")
    else:
        print(f"\nError: No subjects found for student '{student_name_input}'.")
else:
    print(f"\nError: Student '{student_name_input}' not found in the records.")
```
