**Python program for multiple-choice questions using loops and if-elif-else conditions. You can modify the questions list to add your own questions.**

```python
# Define a list of dictionaries, where each dictionary represents a question
questions = [
    {
        "question": "What is the capital of Pakistan?",
        "options": ["A) Berlin", "B) Madrid", "C) Islamabad", "D) Rome"],
        "answer": "C"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["A) Earth", "B) Mars", "C) Jupiter", "D) Venus"],
        "answer": "B"
    },
    {
        "question": "What is 2 + 2?",
        "options": ["A) 3", "B) 4", "C) 5", "D) 6"],
        "answer": "B"
    }
]
```

The input validation occurs in the else block of the if-elif-else statement. After the user enters their answer, the code checks if the user_answer matches the q['answer'] (correct) or if it's one of the valid options ('A', 'B', 'C', 'D') but incorrect. If neither of these conditions is met, it means the user's input was not a valid option, and the else block is executed:

    else:
        print("Invalid input. Please enter A, B, C, or D.")
This prints an error message, informing the user that their input was not valid. However, it currently doesn't prompt the user to re-enter their answer for the same question; it simply moves on to the next question.

```python
# Initialize score
score = 0

print("Welcome to the Python Quiz!")
print("-------------------------")

# Loop through each question
for i, q in enumerate(questions):
    print(f"\nQuestion {i + 1}: {q['question']}")
    for option in q['options']:
        print(option)

    user_answer = input("Enter your answer (A, B, C, or D): ").strip().upper()

    # Check if the answer is correct using if-elif-else
    if user_answer == q['answer']:
        print("Correct!")
        score += 1
    elif user_answer in ["A", "B", "C", "D"]:
        print(f"Incorrect. The correct answer was {q['answer']}.")
    else:
        print("Invalid input. Please enter A, B, C, or D.")

print("\n-------------------------")
print(f"Quiz finished! Your final score is {score} out of {len(questions)}.")
```
