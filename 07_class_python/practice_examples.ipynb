```python
inventory = {
    "apple": {"price": 0.50, "stock": 100},
    "banana": {"price": 0.30, "stock": 150},
    "milk": {"price": 2.00, "stock": 50},
    "bread": {"price": 2.50, "stock": 30}
}

shopping_cart = {}

def display_inventory():
    print("\n--- Available Groceries ---")
    for item, details in inventory.items():
        print(f"{item.capitalize()}: Stock = {details['stock']}, Price = ${details['price']:.2f}")
    print("-------------------------")

def add_to_cart():
    item_name = input("Enter item name to add to cart: ").lower()
    if item_name in inventory:
        try:
            quantity = int(input(f"Enter quantity of {item_name} to add: "))
            if quantity > 0:
                if inventory[item_name]["stock"] >= quantity:
                    shopping_cart[item_name] = shopping_cart.get(item_name, 0) + quantity
                    print(f"Added {quantity} {item_name}(s) to your cart.")
                else:
                    print(f"Insufficient stock. Only {inventory[item_name]['stock']} {item_name}(s) available.")
            else:
                print("Quantity must be greater than zero.")
        except ValueError:
            print("Invalid quantity. Please enter a number.")
    else:
        print("Item not found in inventory.")

def view_cart():
    print("\n--- Your Shopping Cart ---")
    if not shopping_cart:
        print("Your cart is empty.")
        return

    total_bill = 0
    for item, quantity in shopping_cart.items():
        price = inventory[item]["price"]
        item_total = quantity * price
        total_bill += item_total
        print(f"{item.capitalize()}: {quantity} x ${price:.2f} = ${item_total:.2f}")
    print("--------------------------")
    print(f"Subtotal: ${total_bill:.2f}")
    print("--------------------------")

def checkout():
    print("\n--- Checking Out ---")
    if not shopping_cart:
        print("Your cart is empty. Nothing to checkout.")
        return

    total_bill = 0
    for item, quantity in shopping_cart.items():
        price = inventory[item]["price"]
        item_total = quantity * price
        total_bill += item_total
        inventory[item]["stock"] -= quantity  # Deduct from stock

    print(f"Your total bill is: ${total_bill:.2f}")
    shopping_cart.clear()  # Clear the shopping cart
    print("Purchase complete! Your cart has been cleared.")
    print("--------------------")

# The main loop will be updated in the next step
```
